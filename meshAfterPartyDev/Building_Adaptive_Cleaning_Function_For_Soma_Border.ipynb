# New algorithm that will add a soma stretching leg

```python
"""
Pseudocode:
- Given a certain skeleton and the soma touching vertices for each soma it borders

0) Create a graph of the mesh from the vertices and edges
1) Project all skeleton points and soma boundary vertices
2) Find the closest skeleton point to the soma border (for that soma)
Based on the output of the closest skeleton point
3a) If it is a node of degree 1 --> do nothing
3b) If Not endpoint:
    Add an edge from the closest skeleton point coordinate to vertex average of all soma boundaries

-- then just pass to regular skeleton cleaning

"""
```

```python
#building the soma_to_piece_touching_vertices dictionaries that can test with
soma_to_piece_touching_vertices = {0:dict(),
                                  1:dict()}
limb_idx = 1
curr_limb = current_neuron[limb_idx]
for st_dict in curr_limb.all_concept_network_data:
    soma_to_piece_touching_vertices[st_dict["starting_soma"]][limb_idx] = st_dict["touching_soma_vertices"]
    
combined_limb_idx = [5,6]
soma_to_piece_touching_vertices[0][4] = np.concatenate([current_neuron[l_idx].current_touching_soma_vertices for l_idx in combined_limb_idx])

# check that got the right boundaries
# sk.graph_skeleton_and_mesh(other_meshes=[recovered_limb_mesh],
#                           other_scatter=[soma_to_piece_touching_vertices[0][4]],
#                           scatter_size=1)

# sk.graph_skeleton_and_mesh(other_meshes=[current_neuron[1].mesh],
#                           other_scatter=[soma_to_piece_touching_vertices[0][1],soma_to_piece_touching_vertices[1][1]],
#                           scatter_size=1)
```

```python
sk.graph_skeleton_and_mesh(other_meshes=[current_neuron[1].mesh],
                           other_skeletons=[current_neuron[1].skeleton],
                          #other_scatter=[soma_to_piece_touching_vertices[0][1],soma_to_piece_touching_vertices[1][1]],
                          #scatter_size=1
                          )
```

```python
# current_skeleton = deepcopy(new_skeleton)
# skeleton_mesh = recovered_limb_mesh

current_skeleton = deepcopy(current_neuron[1].skeleton)
skeleton_mesh = current_neuron[1].mesh
```

```python
from pykdtree.kdtree import KDTree
import networkx_utils as xu
xu = reload(xu)
z = 1
for sm,v in soma_to_piece_touching_vertices.items():
    if z not in v:
        continue

    #0) Create a graph of the mesh from the vertices and edges and a KDTree
    vertex_graph = tu.mesh_vertex_graph(skeleton_mesh)
    mesh_KD = KDTree(skeleton_mesh.vertices)

    #1)  Project all skeleton points and soma boundary vertices onto the mesh
    all_skeleton_points = np.unique(current_skeleton.reshape(-1,3),axis=0)
    sk_points_distances,sk_points_closest_nodes = mesh_KD.query(all_skeleton_points)
    
    sbv = soma_to_piece_touching_vertices[sm][z]
    soma_border_distances,soma_border_closest_nodes = mesh_KD.query(sbv[0].reshape(-1,3))
    
    #2) Find the closest skeleton point to the soma border (for that soma), find shortest path from many to many
    path,closest_sk_point,closest_soma_border_point = xu.shortest_path_between_two_sets_of_nodes(vertex_graph,sk_points_closest_nodes,soma_border_closest_nodes)
    
    #3) Find closest skeleton point
    closest_sk_pt = np.where(sk_points_closest_nodes==closest_sk_point)[0][0]
    closest_sk_pt_coord = all_skeleton_points[closest_sk_pt]
    sk_graph = sk.convert_skeleton_to_graph(current_skeleton)
    #find the node that has the desired vertices and its' degree
    sk_node = xu.get_nodes_with_attributes_dict(sk_graph,dict(coordinates=closest_sk_pt_coord))[0]
    sk_node_degree = sk_graph.degree()[sk_node]
    
    if sk_node_degree == 0:
        raise Exception("Found 0 degree node in skeleton")
    elif sk_node_degree == 1: #3a) If it is a node of degree 1 --> do nothing
        print(f"skipping soma {sm} because closest skeleton node was already end node")
        continue
    else:
        #3b) If Not endpoint:
        #Add an edge from the closest skeleton point coordinate to vertex average of all soma boundaries
        print("Adding new branch to skeleton")
        new_branch_sk = np.vstack([closest_sk_pt_coord,np.mean(sbv,axis=0)]).reshape(-1,2,3)
        current_skeleton = sk.stack_skeletons([current_skeleton,new_branch_sk])

```

```python
from pykdtree.kdtree import KDTree
import networkx as nx
import trimesh_utils as tu
tu = reload(tu)

new_G = sk.convert_skeleton_to_graph(new_cleaned_skeleton)
end_nodes = np.array([k for k,n in dict(new_G.degree()).items() if n == 1])
skeleton_mesh = curr_limb.mesh

distance_to_ignore_end_nodes_close_to_soma_border=10000
end_nodes,len(end_nodes)
```

```python
"""
Pseudocode:
0) Get the mesh face midpoints and create a KDTree from them
1) Map the soma border vertices to the face midpoints
2) Map the endpoints to closest face midpoints
3) Create Graph from face adjacency
4) For each endpoint, find shortest distance from endpoint to a soma border along graph
5) Only keep those endpoints whose path is beyond certain distance

"""
print(f"Number of end_nodes BEFORE filtering = {len(end_nodes)}")

end_nodes_coordinates = xu.get_node_attributes(new_G,node_list=end_nodes)
soma_border_vertices = curr_limb.current_touching_soma_vertices

#0) Get the mesh vertices and create a KDTree from them
mesh_KD = KDTree(skeleton_mesh.vertices)

#1) Map the soma border vertices to the mesh vertices
soma_border_distances,soma_border_closest_nodes = mesh_KD.query(soma_border_vertices[0].reshape(-1,3))

#2) Map the endpoints to closest face midpoints
end_nodes_distances,end_nodes_closest_nodes = mesh_KD.query(end_nodes_coordinates)

#3) Create Weighted Graph from vertex edges
vertex_graph = tu.mesh_vertex_graph(skeleton_mesh)

#4) For each endpoint, find shortest distance from endpoint to a soma border along graph
# for en,en_mesh_vertex in zip(end_nodes,end_nodes_closest_nodes):
#     #find the shortest path to a soma border vertex

node_idx_to_eliminate = []
node_idx_to_eliminate_len = []
for en_idx,en in enumerate(end_nodes_closest_nodes):
    try:
        path_len, path = nx.single_source_dijkstra(vertex_graph,
                                                   source = en,
                                                   target=soma_border_closest_nodes[0],
                                                   cutoff=distance_to_ignore_end_nodes_close_to_soma_border
                                                  )
    except:
        node_idx_to_keep.append(en_idx)
    else: #a valid path was found
        node_idx_to_eliminate.append(en_idx)
        node_idx_to_eliminate_len.append(path_len)
        print(f"May Eliminate end_node {en_idx}: {end_nodes[en_idx]} because path_len to soma border was {path_len}")

single_node_idx_to_eliminate = node_idx_to_eliminate[np.argmin(node_idx_to_eliminate_len)]
print(f"single_node_to_eliminate = {single_node_idx_to_eliminate}")
new_end_nodes = np.array([k for i,k in enumerate(end_nodes) if i != single_node_idx_to_eliminate])
print(f"Number of end_nodes AFTER filtering = {len(new_end_nodes)}")
```
