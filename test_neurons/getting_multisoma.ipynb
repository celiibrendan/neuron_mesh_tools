```python
# Download the mesh of a desired soma
```

```python
def segment_id_to_mesh(seg_ID):
    curr_mesh = (minnie.Decimation() & f"segment_id={seg_ID}").fetch1("mesh")
    example_mesh = trimesh.Trimesh(vertices=curr_mesh.vertices,
                               faces=curr_mesh.faces)
    return example_mesh

chosen_segment = 88112389022896485
chosen_segment_mesh = segment_id_to_mesh(chosen_segment)
```

```python
chosen_segment_mesh.export(f"./multi_soma/{chosen_segment}_5_soma.off")
print("hi")
```
