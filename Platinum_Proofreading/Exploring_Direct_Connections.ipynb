# Finding the direct connections 

```python
# total synapses = 65490708
```

```python
possible_presyn_synapses = (minnie.Synapse.proj(segment_id="presyn") & single_soma_table.proj())
possible_postsyn_synapses = (minnie.Synapse.proj(segment_id="postsyn") & single_soma_table.proj())
len(possible_presyn_synapses),len(possible_postsyn_synapses)
```

```python
direct_connections = minnie.Synapse() & possible_presyn_synapses.proj() & possible_postsyn_synapses.proj()
direct_connections_no_self = (direct_connections & "presyn!=postsyn")
direct_connections_no_self
```

```python
total_direct_connections = 1129767
```

# getting the distribution of distances from soma centers

```python
direct_connections_postsyn = direct_connections_no_self.fetch("postsyn")
```

```python
direct_connections_presyn = direct_connections_no_self.fetch("presyn")
direct_connections_centers = direct_connections_no_self.fetch("centroid_x","centroid_y","centroid_z")
direct_connections_coordinates = np.array(direct_connections_centers).T
```

```python
non_self_connection_mask = direct_connections_presyn != direct_connections_postsyn
np.sum(non_self_connection_mask)
```

```python
np.savez("direct_connection_data.npz",
        direct_connections_postsyn=direct_connections_postsyn,
        direct_connections_presyn=direct_connections_presyn,
        direct_connections_coordinates=direct_connections_coordinates,
        single_soma_coordinates=single_soma_coordinates)
```

```python
"""
To find distance from the somas: 
1) Create a lookup dictionary for the postsyn to soma center
2) Iterate through all of the direct connection presyns
  - get the soma center for that presyn
- find the norm distance and save to list
"""
```

```python
import numpy as np

print((len(np.unique(single_soma_ids)),len(single_soma_ids)))

#creating dictionary
soma_center_dict = dict()
for seg_id,soma_coord in zip(single_soma_ids,single_soma_coordinates):
    soma_center_dict[seg_id] = soma_coord
    
total_soma_to_synapse_distance = []
for curr_id,curr_coord in zip(direct_connections_presyn,direct_connections_coordinates):
    curr_distance = np.linalg.norm(curr_coord-soma_center_dict[curr_id])
    total_soma_to_synapse_distance.append(curr_distance)
```

```python
from matplotlib import pyplot as plt
fig,ax = plt.subplots(1,1)
soma_to_synapse_bins = plt.hist(np.array(total_soma_to_synapse_distance)/1000,bins=100)
ax.set_xlabel("distance from presyn synapse to soma (um)")
ax.set_ylabel("frequency")
ax.set_title("Distance from synapse to presyn soma frequency chart")
plt.show()
```

# getting the soma to soma distance

```python
total_soma_to_soma_distance = []
for curr_presyn,curr_postsyn in zip(direct_connections_presyn,direct_connections_postsyn):
    curr_distance = np.linalg.norm(soma_center_dict[curr_presyn]-soma_center_dict[curr_postsyn])
    total_soma_to_soma_distance.append(curr_distance)
```

```python
from matplotlib import pyplot as plt
fig,ax = plt.subplots(1,1)
soma_to_soma_bins = plt.hist(np.array(total_soma_to_soma_distance)/1000,bins=100)
ax.set_xlabel("Distance from presyn to postsyn soma (um)")
ax.set_ylabel("frequency")
ax.set_title("Distance from presyn to postsyn soma frequency chart")
plt.show()
```

# finding those that are greater than 100 um
