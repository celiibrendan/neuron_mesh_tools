# Viewing the Somas

```python
curr_segment = 79183941617083896
soma_vertices, soma_faces = (minnie.BaylorSegmentCentroid() & dict(segment_id=curr_segment)).fetch("soma_vertices","soma_faces")
from pathlib import Path
my_folder = Path(f"./{curr_segment}_test")
my_folder.mkdir(exist_ok=True)
```

```python
import trimesh
total_soma_mesh = trimesh.Trimesh(vertices=np.array([]),faces=np.array([]))
for verts,faces in zip(soma_vertices,soma_faces):
    curr_mesh = trimesh.Trimesh(vertices=verts,faces=faces)
    total_soma_mesh += curr_mesh
total_soma_mesh.show()
```

```python
#export the mesh into its own folder

soma_file = my_folder / f"{curr_segment}_total_somas.off"
total_soma_mesh.export(f"{str(soma_file)}")
print("hello")
```
